Now, we are interested in how age affects blood pressure i.e. we may try to understad blood pressure as a function of age. This can modelled as linear relationship between blood pressure and age as follow:
blood_pressure = slope * age + intercept
We can estimate the unknowns Slope and Intercept and plot:

```python
import numpy as np
```

```python
Slope, Intercept = np.polyfit(heart_df["Age"], heart_df["Blood Pressure"],1)
print("Slope",Slope)
print("Intercept",Intercept)
```

Now we have estimated Slope and intercept, now we plot regression straight line.

```python
plt.plot(heart_df["Age"],heart_df["Blood Pressure"],"o")
plt.plot(heart_df["Age"],Slope + heart_df["Age"] + Intercept)
plt.xlabel("Age",fontsize = 15)
plt.ylabel("Blood Pressure",fontsize = 15)
```
